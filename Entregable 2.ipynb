```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('ENTREGABLE1.csv')
df
```

```python
# Creacion del grafico de barras 
fig, ax = plt.subplots()
ax.hist(df['SEGMENT'], bins = 10)
ax.set_title('Frecuencia de las categorÃ­as de SEGMENTO')
ax.set_xlabel('Segmento')
ax.set_ylabel('Frecuencia')
plt.xticks(rotation=45) # Rotacion de etiquetas del eje x
plt.show
```

```python
fig.savefig('Frecuencia de segmento.png')
```

```python
# Contar las ocurrencias de cada categorÃ­a
segment_counts = df['SEGMENT'].value_counts()

# Crear un grÃ¡fico de barras
fig, ax = plt.subplots()
ax.bar(segment_counts.index, segment_counts.values)
ax.set_title('Frecuencia de las categorÃ­as de SEGMENTO')
ax.set_xlabel('Segmento')
ax.set_ylabel('Frecuencia')
plt.xticks(rotation=45)  # Rotar las etiquetas del eje x si es necesario
plt.show()
```

```python
sns.boxplot(data = df, y = 'SEGMENT', x = 'TOTAL_UNIT_SALES')
```

```python
sns.boxplot(data = df, y = 'SEGMENT', x = 'TOTAL_SALES')
```

OBSERVACION 
* Bleach, liquid & gel son los segmentos con mayor frecuencia seguido de pretreat.
* Powder, sanitizer y bar son los segmentos con menor frecuencia
* En principio no se observan diferencias en ventas unitarias y ventas totales entre los segmentos con mayor frecuencia.
* Pretreat teniendo una frecuencia mucho menor que los segmentos Bleach y liquid & gel tiene ventas parecidas

```python
# La columna 'DATE' la hacemos tipo datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Ordenamos los datos por fecha
df = df.sort_values('DATE') 

# Creamos el grafico
df.set_index('DATE')['TOTAL_SALES'].plot(figsize=(10, 5), title="Ventas Totales a lo Largo del Tiempo")
plt.xlabel("Fecha")
plt.ylabel("Ventas Totales")
plt.grid()
plt.show()
```

```python
# Convertir la columna de fechas (por ejemplo, 'DATE') a formato datetime
df['DATE'] = pd.to_datetime(df['DATE'])

# Ordenar los datos por fecha
df = df.sort_values('DATE')

# Crear la grÃ¡fica de lÃ­nea
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='DATE', y='TOTAL_UNIT_SALES', label='Ventas Totales', color='blue')

# Personalizar la grÃ¡fica
plt.title('Ventas Totales a lo Largo del Tiempo', fontsize=16)
plt.xlabel('Fecha', fontsize=12)
plt.ylabel('Ventas Totales', fontsize=12)
plt.xticks(rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

```python
fig.savefig('VT a lo largo del tiempo.png')
```

OBSERVACION 
* Se observa una venta ciclica a lo largo del tiempo con un aumento de ventas en la segunda mitad del 2023

```python
fig, ax = plt.subplots()
monthly_sales = df.groupby(pd.Grouper(key='DATE', freq='M'))['TOTAL_SALES'].sum().reset_index()
ax.scatter(monthly_sales['DATE'], monthly_sales['TOTAL_SALES'])
ax.set_title('Ventas Totales Agrupadas por Mes', fontsize=16)
ax.set_xlabel('Mes', fontsize=12)
ax.set_ylabel('Ventas Totales', fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Agrupar por SEGMENT y sumar TOTAL_SALES
grouped_data = df.groupby('SEGMENT')['TOTAL_SALES'].sum().reset_index()

# Crear el grÃ¡fico de dispersiÃ³n
fig, ax = plt.subplots()
ax.scatter(grouped_data['SEGMENT'], grouped_data['TOTAL_SALES'], color='blue', alpha=0.7)

# PersonalizaciÃ³n del grÃ¡fico
ax.set_title('DispersiÃ³n de Ventas Totales por Segmento (Agrupado)', fontsize=16)
ax.set_xlabel('Segmento', fontsize=12)
ax.set_ylabel('Ventas Totales', fontsize=12)

plt.tight_layout()
plt.show()
```

OBSERVACION
* Las ventas totales estan correlacionadas a las fechas, dando a entender ventas ciclicas o de temporada

```python
# Agrupar las ventas por segmento
sales_by_segment = df.groupby('SEGMENT')['TOTAL_SALES'].sum().sort_values(ascending=False)

# Crear el grÃ¡fico de barras
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(sales_by_segment.index, sales_by_segment.values)

# Configurar el grÃ¡fico
ax.set_title('Ventas Totales por Segmento', fontsize=16)
ax.set_xlabel('Segmento', fontsize=12)
ax.set_ylabel('Ventas Totales', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Mostrar el grÃ¡fico
plt.show()
```

```python
fig.savefig('VT por segmento.png')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Agrupar por regiÃ³n y segmento, sumando las ventas
sales_grouped = df.groupby(['REGION', 'SEGMENT'])['TOTAL_VALUE_SALES'].sum().unstack()


sales_grouped.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Ventas Totales por RegiÃ³n y Segmento', fontsize=16)
plt.xlabel('RegiÃ³n', fontsize=12)
plt.ylabel('Ventas Totales (Valor)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Segmento', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
fig.savefig('VT por segmento y region.png')
```
