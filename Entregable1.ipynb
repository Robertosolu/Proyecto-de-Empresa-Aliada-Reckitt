```python
#importamos archivo de productos y revisamos
producto = pd.read_excel("DIM_PRODUCT.xlsx")
producto
```

```python
producto.info()
```

```python
producto.describe()
```

```python
# Eliminamos valores nulos
producto1 = producto
producto1.dropna(inplace = True)
producto1
```

```python
# Eliminamos filas duplicadas
df_producto = producto1
df_producto.drop_duplicates()
df_producto
```

```python
# Revision de duplicados en columnas
columnas = ['MANUFACTURER', 'BRAND', 'ITEM', 'ITEM_DESCRIPTION', 'FORMAT', 'ATTR1', 'ATTR2', 'ATTR3', 'CATEGORY']
for columna in columnas:
    print(f'columna {columna}: {df_producto[columna].nunique()} opciones')
```

```python
del df_producto['CATEGORY']
df_producto
```

```python
df_producto = df_producto.rename(columns = {'ITEM' : 'ITEM_CODE'})
df_producto
```

```python
# Importamos archivo de segmentos y revisamos
segmento = pd.read_excel("DIM_SEGMENT.xlsx")
segmento
```

```python
segmento.info()
```

```python
segmento.describe()
```

```python
# Eliminamos valores nulos
segmento1 = segmento
segmento1.dropna(inplace = True)
segmento1
```

```python
# Eliminamos filas duplicadas
df_segmento = segmento1
df_segmento.drop_duplicates()
df_segmento
```

```python
# Revision de duplicados en columnas
columnas = ['ATTR1', 'ATTR2', 'ATTR3', 'FORMAT', 'SEGMENT', 'CATEGORY']
for columna in columnas:
    print(f'columna {columna}: {df_segmento[columna].nunique()} opciones')
```

```python
del df_segmento['CATEGORY']
df_segmento
```

```python
# Unir los Dataframe relevantes

df_ventas_calendario = pd.merge(ventas, calendario, on = 'WEEK', how = 'inner')
df_ventas_calendario
```

```python
df_producto_df_segmento = pd.merge(df_producto, df_segmento, on = 'FORMAT', how = 'inner')
df_producto_df_segmento

df_producto_df_segmento = df_producto_df_segmento.drop(columns = ['ATTR2_y', 'ATTR3_y', 'ATTR1_x'])
df_producto_df_segmento
```

```python
df_final = pd.merge(df_ventas_calendario, df_producto_df_segmento, on = 'ITEM_CODE', how = 'inner')
df_final
```

```python
df_final['DATE'] = pd.to_datetime(df_final["DATE"], errors= 'coerce', dayfirst = True)
df_final
```

```python
# Adicion de una nueva columna 
df_final['TOTAL_SALES'] = df_final['TOTAL_UNIT_SALES'] * df_final['TOTAL_VALUE_SALES']
df_final
```

```python
df_final.to_csv('ENTREGABLE1.csv', index = False)
print('El archivo csv se ha guardado correctamente.')
```
